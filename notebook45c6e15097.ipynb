```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# Load Titanic
df = pd.read_csv('/kaggle/input/competitions/titanic/train.csv')
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Fare"] = df["Fare"].fillna(df["Fare"].median())
df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1

features = ["Pclass", "Sex", "Age", "Fare", "FamilySize"]
X = df[features]
y = df["Survived"]

# IMPORTANT: Scale features for SVM!
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train SVM
svm_model = SVC(kernel='rbf', random_state=42)

# Cross Validation
scores = cross_val_score(svm_model, X_scaled, y, cv=5)
print("SVM Scores:", scores.round(3))
print("Mean:", scores.mean().round(3))
print("Std:", scores.std().round(3))
```

```python
# Compare different SVM kernels
kernels = ['linear', 'rbf', 'poly']

for kernel in kernels:
    svm = SVC(kernel=kernel, random_state=42)
    scores = cross_val_score(svm, X_scaled, y, cv=5)
    print(f"Kernel: {kernel}")
    print(f"Mean: {scores.mean().round(3)}")
    print(f"Std: {scores.std().round(3)}")
    print()
```
