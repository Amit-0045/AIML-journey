```python
import numpy as np
import matplotlib.pyplot as plt

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

x = np.linspace(-5, 5, 100)

plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.plot(x, sigmoid(x), color='blue')
plt.title("Sigmoid")
plt.grid(True)

plt.subplot(1,3,2)
plt.plot(x, relu(x), color='red')
plt.title("ReLU")
plt.grid(True)

plt.subplot(1,3,3)
plt.plot(x, tanh(x), color='green')
plt.title("Tanh")
plt.grid(True)

plt.show()
```
