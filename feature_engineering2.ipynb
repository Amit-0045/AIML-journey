```python
import pandas as pd

df = pd.read_csv('/kaggle/input/competitions/titanic/train.csv')

# One Hot Encoding on Embarked column
embarked_encoded = pd.get_dummies(df["Embarked"], prefix="Embarked")
print(embarked_encoded.head())
```

```python
embarked_encoded = pd.get_dummies(df["Embarked"], prefix="Embarked").astype(int)
print(embarked_encoded.head())
```

```python
# Add encoded columns to df
df = pd.concat([df, embarked_encoded], axis=1)

# Drop original Embarked column (no longer needed)
df.drop("Embarked", axis=1, inplace=True)

print(df.head())
print(df.shape)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Method 1 - StandardScaler
# Converts to mean=0, std=1
scaler = StandardScaler()
df["Age_scaled"] = scaler.fit_transform(df[["Age"]])
print(df["Age_scaled"].head())

# Method 2 - MinMaxScaler  
# Converts to range 0-1
minmax = MinMaxScaler()
df["Fare_scaled"] = minmax.fit_transform(df[["Fare"]])
print(df["Fare_scaled"].head())
```
