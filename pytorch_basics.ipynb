```python
import torch

# Create tensors
t1 = torch.tensor([1.0, 2.0, 3.0])
t2 = torch.tensor([4.0, 5.0, 6.0])

print("Tensor 1:", t1)
print("Tensor 2:", t2)

# Basic operations
print("Addition:", t1 + t2)
print("Multiplication:", t1 * t2)
print("Mean:", t1.mean())

# 2D tensor
matrix = torch.tensor([[1.0, 2.0],
                        [3.0, 4.0]])
print("Matrix shape:", matrix.shape)
print("Matrix transpose:", matrix.T)

# Convert from NumPy
import numpy as np
numpy_array = np.array([1, 2, 3])
tensor_from_numpy = torch.from_numpy(numpy_array)
print("From NumPy:", tensor_from_numpy)

# Autograd example
x = torch.tensor(3.0, requires_grad=True)
y = x ** 2  # y = x²
y.backward()  # calculate gradient
print("Gradient of x²  at x=3:", x.grad)  # should be 6 (2x)
```
